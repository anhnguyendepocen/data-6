```python
## Why do we randomize?
group_1 = np.random.randint(0, 50, 20)
group_2 = np.random.randint(50, 100, 20)
```

```python
# Systematic differences exist between both groups:
np.mean(group_1)
np.mean(group_2)
np.append(group_1, group_2)
```

```python
# This table combines the two:
combined_groups = Table().with_columns("Numbers", np.append(group_1, group_2))
combined_groups
```

```python
# Now, if we randomly selected 2 groups
group_1_new = combined_groups.sample(20, with_replacement = False).column(0)
np.average(group_1_new)
```

```python
## Ignore this code; we're just finding the values not in group 1
group_2_new = np.setdiff1d(combined_groups.column(0), group_1_new)
np.average(group_2_new)
```
